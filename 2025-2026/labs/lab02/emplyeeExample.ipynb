```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/employees.csv')

# Display the first few rows
print(df.head())
```

```python
import matplotlib.pyplot as plt


plt.hist(df['Salary'])
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Histogram')
plt.show()
```

```python
df1 = df[df['Senior Management'] == True]
plt.hist(df1['Salary'] )
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary for Senior Management - Histogram')
plt.show()
```

```python
df1 = df[df['Team'] == 'Marketing']
plt.hist(df1['Salary'] )
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary for Marketing - Histogram')
plt.show()
```

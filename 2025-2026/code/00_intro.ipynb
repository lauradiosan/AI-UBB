```python
from huggingface_hub import InferenceClient

# Replace 'your_hf_api_token_here' with your actual API token
client = InferenceClient(token="....")
model="meta-llama/Meta-Llama-3-8B-Instruct"

def requestQuery(client, question, oneModel):
    output = client.chat.completions.create(
        model = oneModel,
        messages=[
            {"role": "system", "content": question},
            # {"role": "user", "content": "Why teach a course on building language models from scratch? Answer in one sentence."},
        ],
        stream=True,
        max_tokens=1024,
    )
    answer = ""
    l = len(answer)
    limit = 800
    for chunk in output:
        if chunk.choices:  # Check if choices list is not empty
            if chunk.choices[0].delta.content is not None:
                answer += chunk.choices[0].delta.content
                l = l + len(chunk.choices[0].delta.content)
                if (l >= limit):
                    answer += "\n"
                    l = 0

            
    return answer
```

```python
def motivation():
    print("De ce ne-am gandit la aceasta structura a cursului?")
    print("     Llama (en):", requestQuery(client, "Why teach a course on building AI models from scratch? Answer in one sentence.", model))
    print("     Llama (ro): ", requestQuery(client, "De ce sa predam un curs despre dezvoltarea de la zero a modelelor de AI? Raspunde intr-o singura propozitie.", model))
    print("     Llama (ro): ", requestQuery(client, "De ce sa predam un curs despre dezvoltarea de la zero a modelelor de AI? Raspunde detaliat.", model))
    print("Deci, e un curs despre a intelege AI prin dezvoltare (nu doar prin utilizare)!!!!")
```
